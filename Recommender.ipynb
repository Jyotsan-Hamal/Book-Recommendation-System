#### We will merge the ratings with books dataframe 

```python
book_ratings = ratings.merge(books,on="ISBN")
```

#### we will use groupby operation on Book-title and see the count of number of rating each book has got

```python
book_ratings.groupby("Book-Title").count()
```

## Below code takes the number of book ratings and forms a new dataframe using:
- reset_index()

```python
books_with_no_of_ratings = book_ratings.groupby("Book-Title").count()['Book-Rating'].reset_index()
```

#### Now renaming the dataframe 

```python
books_with_no_of_ratings.rename(columns={"Book-Rating":"num_ratings"},inplace=True)
books_with_no_of_ratings
```

#### Again doing the groupby operation on Book-rating but with mean approach.

```python
books_with_avg_no_of_ratings = book_ratings[["Book-Rating","Book-Title"]].groupby("Book-Title").mean().reset_index()
books_with_avg_no_of_ratings.rename(columns={"Book-Rating":"avg_num_ratings"},inplace=True)
books_with_avg_no_of_ratings
```

## Now let's merge both of them 

```python
popular_book_df = books_with_no_of_ratings.merge(books_with_avg_no_of_ratings,on="Book-Title")
```

## Now we only want books with number of rating more than 250.
- we also sorted the books in descending order on the basis of "avg_num_ratings" columns

```python
popular_book_df = popular_book_df[popular_book_df['num_ratings']>=250].sort_values("avg_num_ratings",ascending=False)
```

```python
popular_book_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[['Book-Title','num_ratings','avg_num_ratings','Book-Author']]
```

## Below lies the collaborative filtering system

#### Now, we will use groupby method on User-ID to take out those users who has voted the books more than 200 times

```python
x = book_ratings.groupby('User-ID').count()['Book-Rating']>200
usersID_who_voted_more_than_200 = x[x].index
```

```python
users_voted_more_than_200_times =  book_ratings[book_ratings['User-ID'].isin(usersID_who_voted_more_than_200)]
```

```python
users_voted_more_than_200_times.head(5)
```

#### Now we will use groupby on Book-title for those books which has been rated more than 50 times by users.

```python
y = users_voted_more_than_200_times.groupby("Book-Title").count()['Book-Rating']>50
index_of_books_which_has_been_rated_more_than_50_times = y[y].index
```

```python
index_of_books_which_has_been_rated_more_than_50_times
```

```python
final_df = users_voted_more_than_200_times[users_voted_more_than_200_times['Book-Title'].isin(index_of_books_which_has_been_rated_more_than_50_times)]
```

```python
final_df.head(3)
```

```python
final_df.duplicated().sum()
```

#### Pivot Table function will make a rows as 'Book-Title' and columns as 'User-ID' and inside those 'Book-Rating'

```python
pivot_table = final_df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot_table.shape
```

#### User = 679 
#### Books = 810

```python
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table
```

### As you can see the book has became vectors of len = 810.
- each books vector is 810 and now we can see cosine similarity between each books

```python
from sklearn.metrics.pairwise import cosine_similarity
```

## Similarity Scores of each 679 books with 679 books

```python
similarity = cosine_similarity(pivot_table)
```

```python
similarity.shape
```

### Now we use similarity matrix for choosing the top 5 closest books
- np.where provides the index of book_name provided by user in our pivot_table
- then we use similarity matrix to find out the similarity scores
- we enumerates the scores because it includes both scores and index position
- then we sort the similarity_scores_with_indexes in descending order
- takeout only top5 books
- suggest them

```python
def recommend(book_name):
    index_position = np.where(pivot_table.index==book_name)[0][0]
    
    similarity_scores_ = similarity[index_position]
    
    similarity_scores_with_indexes = list(enumerate(similarity_scores_))
    reverse_sorted_similarity_scores_with_indexes = sorted(similarity_scores_with_indexes,reverse=True,key=lambda x:x[1])
    top5_books = reverse_sorted_similarity_scores_with_indexes[0:6]
   
    suggestion = []
    for i in top5_books:
        suggestion.append(pivot_table.index[i[0]])
    
    return suggestion
```

```python
recommend("The Two Towers (The Lord of the Rings, Part 2)")
```
