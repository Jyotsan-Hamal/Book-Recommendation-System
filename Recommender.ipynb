## Things we are doing
- we need to find out the avg rating of each books and see which on is most popular 
- we will see the books with avf rating on descending order.

```python
print(ratings['User-ID'].nunique())
print(ratings['ISBN'].nunique())
```

#### We will merge the ratings with books dataframe 

```python
book_ratings = ratings.merge(books,on="ISBN")
```

```python
book_ratings.head(3)
```

#### we will use groupby operation on Book-title and see the count of number of rating each book has got

```python
book_ratings.groupby("Book-Title").count()
```

## Below code takes the number of book ratings and forms a new dataframe using:
- reset_index()

```python
books_with_no_of_ratings = book_ratings.groupby("Book-Title").count()['Book-Rating'].reset_index()
```

#### Now renaming the dataframe 

```python
books_with_no_of_ratings.rename(columns={"Book-Rating":"num_ratings"},inplace=True)
books_with_no_of_ratings
```

#### Again doing the groupby operation on Book-rating but with mean approach.

```python
books_with_avg_no_of_ratings = book_ratings[["Book-Rating","Book-Title"]].groupby("Book-Title").mean().reset_index()
books_with_avg_no_of_ratings.rename(columns={"Book-Rating":"avg_num_ratings"},inplace=True)
books_with_avg_no_of_ratings
```

## Now let's merge both of them 

```python
popular_book_df = books_with_no_of_ratings.merge(books_with_avg_no_of_ratings,on="Book-Title")
```

## Now we only want books with number of rating more than 250.
- we also sorted the books in descending order on the basis of "avg_num_ratings" columns

```python
popular_book_df = popular_book_df[popular_book_df['num_ratings']>=250].sort_values("avg_num_ratings",ascending=False)
```

```python
popular_book_df.head(40)
```

```python
popular_book_df.merge(books,on="Book-Title").drop_duplicates("Book-Title")[['Book-Title','num_ratings','avg_num_ratings','Book-Author']]
```
